# Stacked Bar Plot

## Top 10 Countries with Male and Female Populations (2022)

```python
# Merge male and female population data on 'Country Name'
merged_data = pd.merge(male_population_data, female_population_data, on="Country Name", suffixes=("_male", "_female"))
```

```python
merged_data
```

```python
# Calculate the total population for each country (male + female)
merged_data["Total Population"] = merged_data["2022_male"] + merged_data["2022_female"]
```

```python
merged_data.head()
```

```python
# Sort data based on total population in descending order
sorted_data = merged_data.sort_values(by="Total Population", ascending=False)

# Select the top 10 countries with the highest total population
top_10_countries = sorted_data.head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(
    top_10_countries['Country Name'],
    top_10_countries['2022_female'],
    label='Female'
)

plt.bar(
    top_10_countries['Country Name'],
    top_10_countries['2022_male'],
    bottom=top_10_countries['2022_female'],
    label='Male'
)
plt.title("Top 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Bottom 10 Countries with Male and Female Populations (2022)

```python
# Select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country Name", y="2022_female", data=bottom_10_countries, color="red", label="Female Population")
sns.barplot(x="Country Name", y="2022_male", data=bottom_10_countries, bottom=bottom_10_countries["2022_female"], color="green", label="Male Population")
plt.title("Bottom 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.histplot(
    total_population_data['2022'],
    bins=30,
    kde=True,
    color='royalblue',
    edgecolor='black',
    alpha=0.8
)

plt.xscale('log')

plt.title(
    'Distribution of World Population (2022)',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Population (Log Scale)', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)

plt.grid(alpha=0.3)

plt.show()
```

```python
print(f"Median Population: {total_population_data['2022'].median():,.0f}")
print(f"Mean Population: {total_population_data['2022'].mean():,.0f}")
print(f"Maximum Population: {total_population_data['2022'].max():,.0f}")
print(f"Minimum Population: {total_population_data['2022'].min():,.0f}")
```

```python
plt.figure(figsize=(12,3))

sns.boxplot(
    x=total_population_data['2022'],
    color='mediumseagreen',
    width=0.5,
    flierprops=dict(
        marker='o',
        markerfacecolor='red',
        markersize=8,
        alpha=0.7
    )
)

plt.xscale('log')

plt.title(
    'Population Distribution (2022)',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel('Population (Log Scale)', fontsize=12)

plt.grid(alpha=0.3)

plt.show()
```

```python
years = [str(y) for y in range(2001, 2023)]

plt.figure(figsize=(12,8))

sns.heatmap(
    total_population_data[years].corr(),
    cmap='coolwarm'
)

plt.title("Population Correlation Across Years")
plt.show()
```

# Key Insights

### 1. Population Distribution is Highly Skewed

The histogram shows that most countries have relatively small populations, while only a few countries have extremely large populations. This creates a right-skewed distribution with several high-value outliers.

### 2. India and China Dominate Global Population

The top population charts reveal that India and China have significantly larger populations than all other countries, making them clear outliers in the dataset.

### 3. Most Countries Have Populations Below 100 Million

The majority of countries fall within the lower population ranges, indicating that highly populated nations represent only a small fraction of all countries worldwide.

### 4. Population Rankings Remain Stable Over Time

Comparisons between 2001, 2010, 2016, and 2022 show that the most populous countries consistently remain among the top-ranked nations, although their population sizes continue to grow.

### 5. Male and Female Population Trends are Similar

Countries with the highest total populations also have the highest male and female populations. The distribution pattern remains consistent across both genders.

### 6. Significant Population Disparity Exists

The gap between the most populated and least populated countries is extremely large, highlighting substantial demographic differences across nations.

### 7. Small Island Nations and Territories Have the Lowest Populations

The bottom 10 population charts indicate that many small island countries and territories have very low population counts compared to the global average.

### 8. Outliers are Clearly Visible in the Box Plot

The box plot identifies several countries as outliers due to their exceptionally large populations. These countries significantly influence the overall distribution.

### 9. Population Growth is Evident Across Years

Comparing population figures from different years suggests a general upward trend in population growth for most countries.

### 10. Global Population is Concentrated in a Few Countries

A relatively small number of countries account for a large proportion of the world's population, while the majority contribute only a small share.
